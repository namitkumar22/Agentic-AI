## What are Guardrails

Guardrails are safety mechanism which controls what goes into and comes out of an AI agent.
Basically they ensure Agent:
1. Processes safe and appropriate input
2. Only perform approved actions
3. Only return validated, compliant output

For Example :

Let's say user asked a query "How to Hack a server". Now this query is not safe so it must be flagged out or the agent should tell like "Hey, this is not good".

There are two approach :

1. Deterministic Approach
    - We use Rule based algorithms like regex, keyword meaning
    - Not able to understand semantics
    - No LLM used so cheaper
2. Model Based
    - Basically we will use LLM here, we will give the input to LLM and we decide whether this message is safe or not
    - Preserves Semantic meaning
    - Any kind of violations can be mentioned to LLM so it catches those violations and stop the execution immediately
    - It is costly because LLM execution is costly

We will use LangChain middleware to implement this. Here are six most important Techniques

1. PII MiddleWare
    - Personal Identifiable Information (PII)
    - Inbuilt MiddleWare and it can detect (email id, credit card, IP's, mac address, URLs). Only these things are supported.
    - It uses Techniques like Redact, Masking, Hashing, of Input, Output and Tool call also
        - redact - it will try to keep the email in a REDACTED_EMAIL format
        - mask - it will ad "*" like (***-***-123)
        - hash - it will make a hash value (87ah45e8)
        - block - Raise an exception

2. Human In The Loop MiddleWae
    - It pauses agents before any sensitive tool execution for Humman approval
    - Needed to implement Threads and Checkpointer

3. Before Agent Hook
    - Runs before LLM call
    - Zero cost for blocked request because we are stoping it before reaching the LLM and if its a Blocked request we can END it immediately

4. After Agent Hook
    - Validates final response before user sees it
    - Can replace or mutate unsafe content

5. Layered Guardrails
    - It combines everything above, we can combine in the form of stack and implement it

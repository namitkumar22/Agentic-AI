## Deep Agents

We already know two types of agent

1. Agent
    - Where the LLM makes a Tool call and the Tool call result will be the output

2. ReAct Agent
    - Where the LLM makes a Tool call and the result if again provided back to the LLM for reasoning then the LLM provides the final output.

Though ReAct Agent is a great advancement of normal Agent, but these both comes under Shalllow Agent because they operate in a basic loop without any structured planning, deep reasoning, persistent memory, etc.

Now comes the concept of Deep Agents

The architecture of Deep Agent is completely different it consiste of 4 important properties

1. Planning Tool
    - When there is a query come it will not hit any kind of tool, first there will be planning. Claude Code usually uses ToDo List for planning.

2. Sub Agents
    - Now this ToDo list need to be executed by someone so we create Sub Agents which are responsible for executing tasks present in ToDo list.

3. System Prompt
    - This tells how our agent should behave

4. File System
    - This is persistent memory which accessible to all the Sub Agents, Now these Sub Agents can do the task and save it in this persistent memory

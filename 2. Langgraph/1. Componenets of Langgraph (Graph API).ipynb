## Langgraph
Langgraph is open-source framework designed by the creators of langchain to build complex, highly controllable and stateful AI Agent workflows. While traditional AI flows follow a straight, sequential line, LangGraph represents workflows as graphs (using nodes and edges), enabling models to loop, branch, and collaborate.

## Components of Langgraph (Graph API)


1. Nodes
2. Edges
3. State

A node is an executable unit of work in a LangGraph workflow. It receives the current state, performs a specific task such as invoking an LLM, calling a tool, or processing data, and returns updates that modify the state.

An edge defines the execution path between nodes. It tells LangGraph which node should run next after the current node completes. In other words, edges control the flow of execution through the graph.

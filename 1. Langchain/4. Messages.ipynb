```python
import os
from dotenv import load_dotenv
load_dotenv()
GROQ_API_KEY = os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq

model = ChatGroq(model="llama-3.3-70b-versatile", api_key=GROQ_API_KEY)
```

## Messages

Messages are the fundamental unit of context for models in Langchain. They represent the input and the output of the models, they carry both the content and metadata needed to present the state of conversation with a LLM. Messages usually contains:

- Role - Identifies the message type (system or user)
- Content - Represent the actual content of the message (e.g text, images, audio, etc)
- Metadata - Optional fields such as information, messageID, token usage, etc

Whenever we give input it is treated as HumanMessage
Whatever model produces output it is a AIMessage

```python
model.invoke("What is Machine Learning ?") # <-- This is treated a human message and in output we get a AIMessage
```

## Text Prompts

Text prompts are strings and good for straightforward generation task when we dont need to retain conversational history

```python
model.invoke("What is Langchain")
```

Use text prompts:
1. When minimum complexity needed
2. When we have a single, standalone request
3. When we dont need to maintain conversational history

## Message Prompts

Message Types:
1. System Message - Tells the model how to behave and provide context for interactions
2. Human Message - Represent the user input and interactions with the model
3. AI Message - Represent the output generated by the model, including text content, tool calls, metadata
4. Tool Message - Represent output of the Tool calls

Here is how simple list of messages looks like

```python
from langchain.messages import SystemMessage, AIMessage, HumanMessage

messages = [
    SystemMessage("You are a expert in Poetry"),
    HumanMessage("Write a poem on Artificial Intelligence")
]

response = model.invoke(messages)
print(response)
```

## Messages with Metadata

```python
message = HumanMessage(
    content="Hello",
    name="Alice", # Optional - Identify different users
    id = "msg_123" # Optional - Unique identifier for tracing
)
```

```python
model.invoke([
    message
])
```

Another example

```python
# Creating a AIMessage manually (e.g. for conversational history representation)
ai_msg = AIMessage("2 + 2 = 4")

response = model.invoke([
    SystemMessage("You are a helpful assistant for math queries"),
    HumanMessage("I have some doubt can you help me"),
    ai_msg, # Inserting as it came from a model
    HumanMessage("Can you help me with 2 + 2")
])

print(response.content)
print(response.usage_metadata)
```

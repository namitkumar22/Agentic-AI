## Here we will see how we can integrate different models with langchain

```python
import os
from dotenv import load_dotenv
load_dotenv()

GROQ_API_KEY = os.getenv("GROQ_API_KEY")
OPENAI_API_KEY = "your-open-ai-api-key"
GOOGLE_API_KEY = "your-google-api-key"
```

1. Load OpenAI Model

```python
from langchain.chat_models import init_chat_model

openai_model = init_chat_model(model="gpt-4.1", model_provider="openai", api_key=OPENAI_API_KEY)
openai_model
```

```python
from langchain_openai import ChatOpenAI

openai_model = ChatOpenAI(model="gpt-4.1", api_key=OPENAI_API_KEY)
openai_model
```

2. Load Google GENAI Model

```python
from langchain.chat_models import init_chat_model

google_model = init_chat_model(model="gemini-1.5-pro", model_provider="google_genai", api_key=GOOGLE_API_KEY)
google_model
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

google_model = ChatGoogleGenerativeAI(model="gemini-1.5-pro", api_key=GOOGLE_API_KEY)
google_model
```

2. Load Groq Model

```python
from langchain.chat_models import init_chat_model

groq_model = init_chat_model(model="llama-3.1-8b-instant", model_provider="groq", api_key=GROQ_API_KEY)
groq_model
```

```python
from langchain_groq import ChatGroq

groq_model = ChatGroq(model="llama-3.1-8b-instant", api_key=GROQ_API_KEY)
groq_model
```

```python
groq_model.invoke("Hello").content
```

## Streaming

Streaming improves user experience for longer responses. calling stream() returns an iteratorthat yieelds output chunks as they are produced

```python
groq_model.stream("write a 200 words paragraph on AI")
```

Here we got a generator object and this is of a stream type and we have to like display the output from the stream, for which we use a for loop

```python
for chunk in groq_model.stream("write a 200 words paragraph on AI"):
  print(chunk.text, end="|", flush=True) # end="|" shows what its generating in a chunk we can make it end="" for a flaseless output
```

## Batch

Batch is a collection of independent request to a model which can significantly improve performace and reduce cost, as the processing can be done in parallel

So lets say we have multiple questions, so we can parallely pass them to the model as a list with the help of batch() function

```python
questions = ["What is AI", "What is ML", "What is DL"]

responses = groq_model.batch(questions)

for response in responses:
  print(response)
```

We can add 1 more parameter here which is config parameter in which we tell what is the max number of parallel calls allowed

```python
responses = groq_model.batch(questions,
                             config={
                                 "max_concurrency" : 5
                                 }
                             )
```

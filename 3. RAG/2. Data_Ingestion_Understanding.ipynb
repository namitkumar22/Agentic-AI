## Understanding Data Ingestion or How we import data

What is Langchain Document

Langchain Document is a kind of data structure which saves data in two format :

1. Page Content - (str) Actual Text or content present inside the file
2. Metadata - (dict) Additional information of file like timespamp, number of pages, file name, etc.

Here is how we create a document or how a Document look like in python

```python
from langchain_core.documents import Document

doc = Document(
    page_content="This is the main text content",
    metadata = {
        "source" : "Example.txt",
        "pages" : 1,
        "author" : "Namit Kumar",
        "date_created" : "19-07-2026"
    }
)

doc
```

There are multiple Loader present in the Langchain to load different types of files like pdf, csv, webpages, etc. We can find all the loaders at this documentation :

https://reference.langchain.com/python/langchain-community/document_loaders

<b><u>Loading Text Content Only</b></u>

We have created two text files in "Text_Files" folder inside the "Data" folder

Now lets see how we can load text files with the help of TextLoader from "langchain.document_loaders" or "langchain_community.document_loaders"

"langchain.document_loaders" is deprecated but it still remains to prevent old code from crashing. It may generate a deprecation warning

This TextLoader returns data in Document object by default. It loads one file at a time

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("Data\\Text_Files\\python.txt", encoding="utf-8") # This create a loader object

document = loader.load() # This loads data from the loader object in Document object

print(document)
```

<b><u>DirecotryLoader</b></u>

Now we can use DirectoryLoader to load all the files present in a directory, but we have to make sure that we pass all the required parameters like all loaders in form of list to load those files.

```python
from langchain_community.document_loaders import DirectoryLoader

from langchain_community.document_loaders import TextLoader

dir_loader = DirectoryLoader(
    "Data/Text_Files", # Path of the directory
    glob = "**/*.txt", # Patterns to match all the files we are trying to load
    loader_cls = TextLoader, # Loader class used to load the files
    loader_kwargs = {"encoding" : "utf-8"}, # Parameters required by the Loader class we are using above
    show_progress = True # Shows a progress bar of how much work is done
)

documents = dir_loader.load()
documents
```

As we can se we got two Document object because we had two text files in our directory

In this code this "glob" is a pattern to match all the files we want to load. Here "*.txt" means look only inside the "Text_Files" folder for all text files only but "**/*.txt" means recursively search and look for text files in all folders present inside that directory too. We can make it "**/*" then it will search and load every file present recursively instead of only text file.

But here in "loader_cls" we can not pass a list of Loaders to load different type of files like pdf and text, etc mixed, we will definetly use "**/*" but we will use "UnstructuredFileLoader" as "loader_cls" which handles different types of files automatically but it is a little slow.

The other options include using two DirectoryLoader class, one with TextLoader and one with PyPDFLoader.

We can also do this manually but it will make code complex

Let's add some pdf files and try these different things

I have downloaded two pdf files from internet one on Machine Learning and one on Agentic AI, now let's try to load it with two different methods

```python
# Loading only PDF files

from langchain_community.document_loaders import PyPDFLoader

dir_loader = DirectoryLoader(
    "Data", # Path of the directory
    glob = "**/*.pdf", # Patterns to match all the files we are trying to load
    loader_cls = PyPDFLoader, # Loader class used to load the files
    show_progress = True # Shows a progress bar of how much work is done
)

# No loader_kwargs needed for PyPDFLoader

document = dir_loader.load()
document
```

Now we got multiple so many documents here with so many metadata, now here PDF contains pages so PyPDFLoader created Documents object for each page so we got multiple document objects page wise but in case of TextLoader there is no pages its a straight text file so for one text file we get one Document whether it contains 1000 lines or 1 line.

In case of PyPDFLoader, we can also use a paramtere {mode = "single"} this return whole PDF content in one document instead of pagewise, default value is {mode = "page"} which returns list of Document object page wise

Now let's see how we can use UnstructuredFileLoader which supports loading different file types like pdf, excel, etc and is also used for Structured Parsing we talked earlier. This loader needs a requirement of "unstructured[all-docs]" and "unstructured-inference" library so make sure it is installed in our environment, this will make sure that UnstrucuredFileLoader is capable of loading common file types like PDF, Docs, images, Excel files, etc.

```python
# Loading all files with different types with the help of UnstructuredFileLoader

# from langchain_community.document_loaders import UnstructuredFileLoader

from langchain_unstructured import UnstructuredLoader ## Successor version of above import, above import is deprecated and no longer maintained
from langchain_community.document_loaders import DirectoryLoader

loader = DirectoryLoader(
    "Data",
    glob = "**/*",
    loader_cls = UnstructuredLoader,
    loader_kwargs = {"mode" : "elements"},
    show_progress = True
)

documents = loader.load()
documents
```

```python
print(len(documents))
```

Different "mode" and arguements we can use above in "loader_kwargs" :

1. {"mode" : "single"} - Provides one Document per file

2. {"mode" : "elements"} - Provides Document for each elements like Heading, Bullet Points, Subheading, etc. This helps in maintaining the Semantic structure and this is used for Structured parsing we talked earlier

3. {"strategy" : "fast"} - Mostly affects PDFs and images, common values are
    - "fast" - Fatser, Simpler extraction, Good for normal PDFs
    - "hi_res" - Better for Tables, Multi-column PDFs, scanned documents, images
    - "ocr_only" - Useful if PDFs contains images only

4. {"encoding" : "utf-8"} - Useful for text file

5. {post_processors : [clean]} - Each extracted element passes through our function for post processing after extraction like making all text to lower letters, it is supposed we have defined a "clean function" above. We can pass many functions in the list and they will be executed in the left to right order

We can also use Docling and LlamaParse

- Docling is better and faster than Unstructured but it doesn't output by default in Langchain Document it outputs in Docling Document and it works locally

- LlamaParse is cloud based developed by LlamaIndex, it also returns LlamaIndex Documents which we have to convert in Langchain Document. It is suitable for research paper, invoices, scanned documents, etc

Both preserve structure very well
